# co2 emmissions

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../data/co2_emission.csv")
df.head()
#sort the data
df = df.sort_values(by='Entity', ascending=True)
#only select the data for India between 2010 and 2017
df = df[(df["Entity"] == "India") & (df["Year"] >= 2010)]
df = df.drop(['Code'], axis=1)
# Rename the 'Entity' column to 'Country'
df.rename(columns={'Entity': 'Country'}, inplace=True)
df.rename(columns={'Annual CO₂ emissions (tonnes )': 'emmissionslarge'}, inplace=True)
df.head(10)
```

```python
#converting large numbers into compact form
#add it into a new column
#only run this if you want to cconvert the numbers of co2 emmissions into a non-exponential number.
def human_format(num):
    num = float('{:.3g}'.format(num))
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    return '{}{}'.format('{:f}'.format(num).rstrip('0').rstrip('.'), ['', 'K', 'M', 'B', 'T'][magnitude])

##human_format(2.466765e+09)

df['emmissionslarge'] = df['emmissionslarge'].apply(human_format) # run only once otherwise re run every cell again!
```

```python
df = df.sort_values(by = "Year")
df.head(20)
```

```python
# Plotting the data
plt.plot(df["Year"], df["emmissionslarge"], marker='o', linestyle='-', color='b', label='CO2 Emissions')

# Adding into axes
plt.xlabel("Year")
plt.ylabel("emmissions: CO2")

# Adding a grid
plt.grid(True)

# Adding a title to the plot
plt.title("CO2 Emissions Over the Years")

# Adding a legend
plt.legend()

# Adding a grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Adding a background color for better contrast
plt.gca().set_facecolor('#f0f0f0')

# Adding a border around the plot
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Plot
plt.show()
```

# use correlation coefficent to show relation between change of emmissions and year 

```python
correlation_coefficient = df["Year"].corr(df["emmissionslarge"])
print(f'Correlation coefficient between {"Year"} and {"emmissionslarge"}: {correlation_coefficient}')

plt.show()
```

```python
# Importing libraries 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.metrics import r2_score

# Calculate the regression line using linregress from scipy
slope, intercept, r_value, p_value, std_err = linregress(df["Year"], df["emmissionslarge"])

# Create a scatter plot
plt.scatter(df["Year"], df["emmissionslarge"], label='Data Points')

# Plot the regression line
plt.plot(df["Year"], slope * df["Year"] + intercept, color='red', label='Regression Line')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel("Emmissions")
plt.title('Co2 emmissions with Regression Line')

# Displaying legend
plt.legend()

# Show the plot
plt.show()

correlation_coefficient = df["Year"].corr(df["emmissionslarge"])
print(f'Correlation coefficient between {"Year"} and {"emmissionslarge"}: {correlation_coefficient}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Calculate the regression line using linregress from scipy
slope, intercept, r_value, p_value, std_err = linregress(new["Year"], new["ANNUAL"])

# Create a scatter plot
plt.scatter(new["Year"], new["ANNUAL"],  label='Temperature', color='blue', alpha=0.7)

# Plot the regression line
plt.plot(new["Year"], slope * new["Year"] + intercept, color='red', linestyle='--', label='Temperature Trend')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title('Change of Temperature Over the Years')

# Displaying legend
plt.legend()

# Show the plot
plt.show()

# Print the correlation coefficient
correlation_coefficient = new["Year"].corr(new["ANNUAL"])
print(f'Correlation coefficient between Year and Average Temperature: {correlation_coefficient}')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.metrics import r2_score

# Train Algorithm (Polynomial)
degree = 4
poly_fit = np.poly1d(np.polyfit(new["Year"],new["ANNUAL"], degree))

# Plot data
xx = np.linspace(2010, 2017, 33)
plt.plot(xx, poly_fit(xx), c='r',linestyle='-')
plt.title('Change of temperature')
plt.xlabel('year')
plt.ylabel('average temperature')
plt.axis([2010, 2017, 28, 33])

# Add grid with increased transparency and linestyle
plt.grid(True, linestyle='--', alpha=0.5)
plt.scatter(new["Year"], new["ANNUAL"])
plt.show()


corr, _ = pearsonr(new["Year"], new["ANNUAL"])
print('Pearsons correlation: %.3f' % corr)
# Calculate and print R-squared score
r2 = r2_score(new["ANNUAL"], poly_fit(new["Year"]))
print(f'R-squared score: {r2:.3f}')
```
